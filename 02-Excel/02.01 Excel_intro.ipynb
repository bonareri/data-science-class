- Excel organizes data in columns and rows and allows you to do mathematical functions.

Excel is typically used for:

- Analysis
- Data entry
- Data management
- Accounting
- Budgeting
- Data analysis
- Visuals and graphs

### 3. Using Filters in Excel

#### What is a Filter?
Filters allow you to view only the rows that match certain criteria, hiding the rest.

#### How to Apply a Filter:

1. Select the header row of your table.
2. Go to the `Data` tab → Click Filter (or press `Ctrl + Shift + L`).
3. Drop-down arrows will appear next to each column.

####  Clear or Remove Filters:
- `Data` tab → Click Clear to reset filters.
- Click Filter again to turn them off completely.

#### Turn Your Data into a Table
- Select your dataset → Press `Ctrl + T`
- This auto-applies filters, formatting, and enables easier sorting and analysis.

### 4. Moving or Copying Columns in Excel
#### A. Move a Column (Drag and Drop)

1. Select the entire column by clicking the column letter (e.g., B).
2. Hover over the border of the selection until the cursor changes to a 4-arrow cross.
3. Hold down Shift, then drag the column to the new location.
4. A green line will show where the column will be dropped.

Release the mouse.

#### B. Copy a Column
1. Select the column.
2. Press Ctrl + C (or right-click → Copy).
3. Select the column where you want to paste.
4. Press Ctrl + V to paste.

#### C. Split columns Use "Text to Columns"
When to use: When values are separated by commas, spaces, or other delimiters.

Steps:

1. Select the column with the combined data (e.g. City, State).
2. Go to Data → Text to Columns.
3. Choose Delimited, then click Next.
4. Choose the delimiter (e.g. comma or space), then Finish.

### 5. Freezing Panes in Excel
Freezing panes helps lock headers or labels so you can scroll and still see key info.

#### A. Freeze Top Row

1. Go to `View` tab.
2. Click `Freeze Panes` → Choose Freeze Top Row.

#### B. Freeze First Column
Use this for wide sheets where you need to keep names or IDs visible.

1. Go to `View` tab.
2. Click `Freeze Panes` → Choose Freeze First Column.

#### C. Freeze Panes Based on Your Selection
To freeze both row and column (e.g., keep headers and labels visible):

1. Click on the cell below the row and to the right of the column you want to freeze.
Example: Click `B2` to freeze row 1 and column A.

2. Go to `View` → Click Freeze Panes → Choose Freeze Panes.

#### D. Unfreeze Panes
To remove the freeze:

- Go to View tab → Click Freeze Panes → Choose Unfreeze Panes



## Data Cleaning in Excel

Cleaning your data in Excel is essential for accuracy and analysis. Below is a structured guide to common cleaning tasks with formulas and tools.

### 1. Remove Duplicates

**Steps:**
1. Select your data range.
2. Go to **Data > Remove Duplicates**.
3. Choose columns to check for duplicates (e.g., Email, ID).

---

### 2. Trim Extra Spaces

Sometimes cells contain leading/trailing spaces that break filters or lookups.

**Formula:**
```excel
=TRIM(A2)
```

### 3. Fix Text Case
| Goal                 | Formula       |
| -------------------- | ------------- |
| ALL CAPS             | `=UPPER(A2)`  |
| lowercase            | `=LOWER(A2)`  |
| Capitalize Each Word | `=PROPER(A2)` |

### 4. Split Data into Columns

Example: Split "Full Name" into "First" and "Last" names.

Steps:

1. Select the column.
2. Go to Data > Text to Columns.
3. Choose Delimited (e.g., space or comma).
4. Follow the wizard.

### 5. Handle Missing Data

a) Filter for Blanks
Use Filter → Check (Blanks) under any column.

b) Replace Blanks with Default

```excel
=IF(A2="", "Unknown", A2)
```

c) Handle Errors
```excel
=IFERROR(formula, "Fix this")
```

### 6. Find and Replace
Fix inconsistent values (e.g., “Nairobii” → “Nairobi”).

Steps:

1. Press Ctrl + H or go to Home > Find & Replace.
2. Enter wrong value, then correct value.
3. Click Replace All.

### 7. Standardize Data Formats
a) Dates
Format: Format Cells > Date

b) Numbers
Format: Format Cells > Number (remove commas, currency symbols)

### 8. Use Data Validation
Prevent invalid entries with dropdowns or rules.

Steps:

1. Select the cell range.
2. Go to Data > Data Validation.
3. Choose from list, number range, date, etc.

### 9. Conditional Formatting
Highlight issues visually.

Steps:

1. Select data.
2. Go to Home > Conditional Formatting.
3. Use rules like:
    - Duplicates
    - Greater/Less than
    - Custom formulas

### 10. Remove Blank Rows or Columns
Steps:

1. Select your data.
2. Go to Home > Find & Select > Go To Special.
3. Choose Blanks → Right-click → Delete.
