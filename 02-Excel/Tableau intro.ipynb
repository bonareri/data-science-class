## Introduction to Tableau

**Tableau** is a powerful data visualization tool used to convert raw data into understandable dashboards and visual reports. It helps users see and understand their data using a drag-and-drop interface for fast analytics.

### Why Use Tableau?
- Easy to use (drag-and-drop)
- Connects to various data sources
- Builds interactive dashboards
- Supports real-time data updates
- Ideal for business intelligence and storytelling

---

## Dimensions and Measures in Tableau

## 1. Dimensions
- **Definition**: Categorical fields that describe and segment data. They are qualitative and usually *non-numeric*.
- **Purpose**: Used to break data into groups (e.g., by Region, Product, Date).
- **Examples**:
  - Product Name
  - Region
  - Order Date
  - Customer ID
  - Category

### Key Characteristics:
- Define the **level of detail** or **granularity** in a view.
- Appear as **blue pills** in Tableau by default.
- Typically **discrete** (distinct categories).
- Placed in rows, columns, filters, color, labels, etc.

---

## 2. Measures
- **Definition**: Numeric fields used for calculation and analysis. They are quantitative and *can be aggregated*.
- **Purpose**: Provide values to be measured (e.g., total sales, average profit).
- **Examples**:
  - Sales
  - Profit
  - Quantity
  - Discount

### Key Characteristics:
- Can be aggregated (SUM, AVG, COUNT, etc.)
- Appear as **green pills** in Tableau by default.
- Typically **continuous** (range of numeric values).
- Used on axes, tooltips, size, and color shelves.

---

## Dimensions vs Measures: Tableau Classification
Tableau automatically classifies fields based on their **data type**:
- **Strings** (text) → Dimension
- **Dates** → Dimension (but can be continuous or discrete)
- **Numbers** → Measure

### Convert Between Them
- Right-click a field > Select **"Convert to Dimension"** or **"Convert to Measure"**
- Useful for treating numeric IDs as labels, or categorizing numeric values

---

## Discrete vs Continuous Fields

### Discrete
- Represent **distinct, separate values**
- Appear as **blue pills**
- Create **headers** in views
- Used for categorization and grouping
- Applies to both **dimensions** and **measures**

  **Examples**:
  - Country
  - Customer Segment
  - Order ID
  - Discrete Month (Jan, Feb...)

### Continuous
- Represent a **range of values**
- Appear as **green pills**
- Create **axes** in views
- Used for trends, averages, and ranges
- Usually applied to **measures**

  **Examples**:
  - Sales
  - Profit
  - Continuous Date (timeline)
  - Age

### Quick Tips
- Start with **dimensions** to define your structure.
- Add **measures** to analyze your data.
- Use **discrete** fields for categories and filters.
- Use **continuous** fields for trends and numeric analysis.
