## Data Blending in Tableau

**Data Blending** is a method used in Tableau to combine data from **multiple data sources** that are **not stored in the same database**. It allows you to analyze related data without needing to perform complex ETL operations or database joins.

---

### What is Data Blending?

- A **way to combine data** from two different data sources **within Tableau**.
- Performed **after data is aggregated**, unlike SQL joins (which happen before aggregation).
- Works by linking a **primary data source** (main) with one or more **secondary sources** (supplemental).
- Useful when:
  - Data is stored in **different systems** (e.g., Excel + SQL Server).
  - You canâ€™t perform a join before loading into Tableau.
  - You need a **loose** relationship between datasets.

---

### Key Concepts

| Term              | Description                                                                 |
|-------------------|-----------------------------------------------------------------------------|
| **Primary Source**| The main data source (fields appear with a blue checkmark).                 |
| **Secondary Source**| The blended data source (fields show orange checkmarks).                |
| **Linking Field** | A common field (like `Date`, `Product ID`, or `Region`) used to connect sources. |

---

**You have:**

### How to Do Data Blending in Tableau

1. **Connect to both data sources** in Tableau.
2. Drag a field from the **primary source** to the view.
3. Then, drag a field from the **secondary source**.
4. Tableau **automatically links** the data if a common field exists (e.g., `Date`, `Product ID`).
   - Youâ€™ll see a **chain icon** in the Data pane (linking field).
5. Use the **linked field** to control how the data sources blend.

### Limitations of Data Blending

- Blending happens **after aggregation**, which may cause mismatched totals.
- Filtering only works well on the **primary source**.
- **One-to-many** blends may not show all rows from the secondary source.
- Performance can be slower with large datasets.

---

### Data Blending vs Joins

| Feature            | Data Blending                    | Joins                           |
|--------------------|----------------------------------|---------------------------------|
| Source Location    | Different data sources           | Same data source/table         |
| Execution Time     | After aggregation (post-join)    | Before aggregation             |
| Link Type          | Common field (soft link)         | Shared key (hard join)         |
| Result Flexibility | More flexible for dashboards     | Better for row-level accuracy  |

---

### Best Practices

- Use **data blending** when you cannot join data before importing to Tableau.
- Ensure linking fields are **clean and matching**.
- Be aware of **aggregation mismatches**.
- Consider **Data Relationships** (newer alternative in recent Tableau versions).
