## Dimensions vs Measures and Discrete vs Continuous in Tableau

Understanding how Tableau categorizes data is key to building effective visualizations. Tableau automatically classifies fields into **Dimensions or Measures**, and identifies them as either **Discrete** or **Continuous**.

---

### Dimensions vs Measures

| Feature     | Dimensions                     | Measures                       |
|-------------|--------------------------------|--------------------------------|
| **Definition** | Qualitative / categorical data | Quantitative / numeric data     |
| **Examples**   | Category, Region, Customer ID | Sales, Profit, Quantity         |
| **Color**      | Blue                          | Green                           |
| **Role**       | Slice or group data            | Perform calculations/analysis   |
| **Aggregation**| Not aggregated                | Can be aggregated (SUM, AVG...)|
| **Placement**  | Rows, Columns, Filters, etc.   | Axes, Size, Tooltip, etc.       |

#### Quick Tip:
- Use **Dimensions** to break down your data.
- Use **Measures** to analyze and quantify it.

---

### How to Convert Between Them

- Right-click a field in the Data Pane.
- Choose **"Convert to Dimension"** or **"Convert to Measure"**.
- This is helpful when a numeric field like `Zip Code` should act like a category.

### Examples

| Field         | Type       | Behavior           |
|---------------|------------|--------------------|
| `Category`    | Dimension  | Discrete (Blue)    |
| `Region`      | Dimension  | Discrete (Blue)    |
| `Order Date`  | Dimension  | Continuous or Discrete |
| `Sales`       | Measure    | Continuous (Green) |
| `Profit`      | Measure    | Continuous (Green) |
| `Age Group`   | Measure    | Discrete (Binned)  |

### Best Practices

- Use **Discrete Dimensions** for labels, filters, and grouping.
- Use **Continuous Measures** for trends, histograms, and timelines.
- Understand how changing **Discrete vs Continuous** affects your chart type and layout.
