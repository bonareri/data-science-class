### Step 1: Import Required Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Step 2: Load a Sample Dataset

```python
# Load 'iris' dataset as an example
df = sns.load_dataset("iris")

# Display first few rows
df.head()
```

### Step 3: Compute a Correlation Matrix

```python
correlation_matrix = df.corr(numeric_only=True)
print(correlation_matrix)
```

### Step 4: Create a Heatmap

```python
plt.figure(figsize=(10,8)) # Optional: set figure size
sns.heatmap(correlation_matrix, 
            annot=True,           # Show values
            cmap="coolwarm",      # Color map
            linewidths=0.5,       # Line between cells
            linecolor='gray')

plt.title("Correlation Heatmap - Iris Dataset")
plt.show()
```

### Key Heatmap Parameters in sns.heatmap()
Parameter	Description
annot=True	Displays numerical values in cells
cmap	Sets the color scheme (e.g., "coolwarm", "viridis")
linewidths	Adds space between cells
linecolor	Color of lines between cells
vmin, vmax	Set min/max values for the color scale
square=True	Make cells square
