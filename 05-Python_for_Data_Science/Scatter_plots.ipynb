```python
df = pd.read_csv('visualizations.csv', encoding='latin1')
df
```

```python
df.set_index('Country')
```

```python
plt.scatter(df['Total cases 2023'], df['No. of deaths (2023)'], c='green', marker='o', s=100)
plt.xlabel('Total cases 2023')
plt.ylabel('No. of deaths (2023)')
plt.title('Total Cases vs. No. of deaths of Malaria (2023)')
plt.show()
```

```python
plt.scatter(df['Total cases 2023'], df['No. of deaths (2023)'])
plt.xlabel('Total cases 2023')
plt.ylabel('No. of deaths (2023)')
plt.title('Total Cases vs No. of Deaths (2023)')
plt.show()
```

we can also change the color by putting in the color in the brackets 

(c = 'red')

```python
plt.scatter(df['Total cases 2023'], df['No. of deaths (2023)'], c='green')
plt.xlabel('Total cases 2023')
plt.ylabel('No. of deaths (2023)')
plt.title('Total Cases vs No. of Deaths (2023)')
plt.show()
```

```python
plt.scatter(df['Total cases 2023'], df['No. of deaths (2023)'], marker='x', c='red')
plt.xlabel('Total cases 2023')
plt.ylabel('No. of deaths (2023)')
plt.title('Total Cases vs No. of Deaths (2023)')
plt.show()
```

we can also change the size

```python
plt.scatter(df['Total cases 2023'], df['No. of deaths (2023)'], s=50)
plt.xlabel('Total cases 2023')
plt.ylabel('No. of deaths (2023)')
plt.title('Total Cases vs No. of Deaths (2023)')
plt.show()
```

```python
df.plot.scatter(x='Total cases 2023', y='No. of deaths (2023)', c='black')
```

```python
import mplcursors

# 'Country', 'Total cases 2023', 'No. of deaths (2023)'

fig, ax = plt.subplots()
sc = ax.scatter(df['Total cases 2023'], df['No. of deaths (2023)'], c='red')

# Labels and title
ax.set_xlabel('Total cases 2023')
ax.set_ylabel('No. of deaths (2023)')
ax.set_title('Total Cases vs No. of Deaths (2023)')

# Enabling hover tooltips
cursor = mplcursors.cursor(sc, hover=True)

@cursor.connect("add")
def on_hover(sel):
    idx = sel.index
    sel.annotation.set(
        text=(
            f"Country: {df['Country'].iloc[idx]}\n"
            f"Total Cases: {df['Total cases 2023'].iloc[idx]}\n"
            f"Deaths: {df['No. of deaths (2023)'].iloc[idx]}"
        ),
        fontsize=9,
        backgroundcolor="white"
    )

plt.show()
```

plt.plot() is faster than scatter with large datasets
