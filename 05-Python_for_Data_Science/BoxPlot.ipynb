A box plot (also called a box-and-whisker plot) in Matplotlib is a standardized way of displaying the distribution of data based on a five-number summary:

Minimum

First quartile (Q1 or 25th percentile)

Median (Q2 or 50th percentile)

Third quartile (Q3 or 75th percentile)

Maximum

It also helps visualize outliers and skewness.

ðŸ§  2.Interpreting a Box Plot
A box plot consists of:

Component	Description
Box	Extends from Q1 to Q3 (the interquartile range, or IQR).
Median line	Line inside the box at the median (Q2).
Whiskers	Extend from Q1 down to the min and Q3 up to the max, excluding outliers.
Outliers	Points outside 1.5 Ã— IQR from Q1/Q3, shown as dots or asterisks.

### When to Use Box Plots
Comparing distributions across multiple categories.

Identifying outliers and variability.

Checking data symmetry or skewness.

Summarizing large data visually.

```python
import matplotlib.pyplot as plt

# Sample data
data = [7, 15, 13, 22, 29, 5, 9, 18, 21, 25]

# Create box plot
plt.boxplot(data)

# Add title and labels
plt.title("Box Plot Example")
plt.ylabel("Values")

# Show plot
plt.show()
```

```python
data = [
    [7, 15, 13, 22, 29],
    [5, 9, 18, 21, 25],
    [6, 8, 14, 19, 24]
]

plt.boxplot(data)
plt.title("Multiple Box Plots")
plt.xticks([1, 2, 3], ["Group 1", "Group 2", "Group 3"])
plt.ylabel("Scores")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Updated data with more points
data = [
    [7, 15, 13, 22, 29, 35, 18, 17, 25, 30],
    [5, 9, 18, 21, 25, 28, 20, 22, 19, 24],
    [6, 8, 14, 19, 24, 27, 26, 20, 23, 21]
]

# Create the boxplot with patch_artist=True to allow coloring
box = plt.boxplot(data, patch_artist=True)

# Colors for each box
colors = ['lightblue', 'lightgreen', 'salmon']

# Apply colors and labels
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

# Set title and labels
plt.title("Multiple Box Plots with Colors")
plt.xticks([1, 2, 3], ["Group 1", "Group 2", "Group 3"])
plt.ylabel("Scores")

# Add custom legend
legend_labels = ["Group 1", "Group 2", "Group 3"]
plt.legend([box["boxes"][i] for i in range(3)], legend_labels, loc="upper left")

plt.show()
```
