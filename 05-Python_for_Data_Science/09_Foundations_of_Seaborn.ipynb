### What is Seaborn?
`Seaborn` is a Python data visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics.

Key Features
| Feature                  | Description                                             |
| ------------------------ | ------------------------------------------------------- |
| **Built-in themes**      | Predefined styling (`whitegrid`, `darkgrid`, etc.)      |
| **DataFrames-friendly**  | Works directly with pandas DataFrames                   |
| **Statistical plotting** | Regression, distribution, box plots, violin plots, etc. |
| **Built-in datasets**    | Datasets like `tips`, `flights`, `iris` for practice    |
| **Easy grouping**        | Supports grouping by hue, size, style                   |


Common Seaborn Plot Types

| Function            | Purpose                                |
| ------------------- | -------------------------------------- |
| `sns.barplot()`     | Categorical data with summary stats    |
| `sns.countplot()`   | Counts of unique values                |
| `sns.boxplot()`     | Distribution and outliers              |
| `sns.violinplot()`  | Distribution with KDE                  |
| `sns.scatterplot()` | Relationship between 2 numerical vars  |
| `sns.lineplot()`    | Trend over time or continuous variable |
| `sns.heatmap()`     | Matrix-style data (correlation, pivot) |


#### What Is Plot Customization?
Plot customization means changing the appearance of a chart to:

- Improve readability,
- Highlight key insights,
- Match your audience’s preferences (or brand),
- Make plots visually appealing.

Key Elements You Can Customize

| Element                   | Description                                   |
| ------------------------- | --------------------------------------------- |
| **Title**                 | Describes what the chart shows.               |
| **Axis Labels**           | Explain what the X and Y axes represent.      |
| **Legend**                | Helps interpret grouped data.                 |
| **Colors**                | Can show group differences or fit your theme. |
| **Markers / Line Styles** | Control how lines or points look.             |
| **Gridlines**             | Make the chart easier to read.                |
| **Size**                  | Adjust for presentations, reports, etc.       |
| **Ticks & Fonts**         | Customize text, angles, spacing.              |

#### How Seaborn + Matplotlib Work Together
- Seaborn simplifies the creation of common statistical plots.
- Matplotlib allows you to customize every visual element.

`Seaborn` creates a plot and returns a `Matplotlib Axes object`, which you can modify further using `matplotlib.pyplot` or `ax` methods.

Scenario: We want to compare the average total bill for each day of the week, broken down by gender.

##### Import Required Libraries


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

- `seaborn` is for statistical plotting.
- `matplotlib.pyplot` gives you more control to customize things like titles, fonts, gridlines, etc.

#####  Load the Dataset

- Loads the tips dataset (included in Seaborn).
- It contains restaurant tips, total bills, gender, time of day, and more.

```python
tips  = sns.load_dataset('tips')
```

```python
tips.info()
```

```python
tips.head()
```

#####  simple bar chart showing average total bill by day.

```python
sns.set_style('darkgrid')
sns.set_palette('muted')
plt.Figure(figsize=(10, 6))
ax = sns.barplot(x = 'day', y = 'total_bill', data = tips,errorbar = None, hue = 'sex')

ax.set_title('Average Total Bill by Day and Gender')
ax.set_xlabel('Day of the week')
ax.set_ylabel('Average Total Bill')
ax.legend(title ='Gender', loc = 'upper left')

plt.show()
```

```python
ax = sns.barplot(data=tips, x="day", y="total_bill")
plt.show()
```

split by gender (Male/Female) using different colors.

##### Set Seaborn Style and Color Palette

```python
sns.set_style("white") # Options: white, dark, darkgrid, ticks
sns.set_palette("pastel")  # Other options: "Set2", "pastel", "deep", "muted"

plt.figure(figsize=(10, 6))
ax = sns.barplot(data=tips, x="day", y="total_bill", errorbar=None, hue="sex")
plt.show()
```

##### Add Title and Axis Labels

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=tips, x="day", y="total_bill", errorbar=None, hue="sex")

ax.set_title("Average Total Bill by Day and Gender")
ax.set_xlabel("Day of the Week")
ax.set_ylabel("Average Total Bill ($)")
plt.show()
```

#####  Customize Font Sizes and Weight

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=tips, x="day", y="total_bill", errorbar=None, hue="sex")

ax.set_title("Average Total Bill by Day and Gender", fontsize=16, fontweight='bold')
ax.set_xlabel("Day of the Week", fontsize=14)
ax.set_ylabel("Average Total Bill ($)", fontsize=14)
plt.show()
```

##### Customize Legend

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=tips, x="day", y="total_bill", errorbar=None, hue="sex")

ax.set_title("Average Total Bill by Day and Gender", fontsize=16, fontweight='bold')
ax.set_xlabel("Day of the Week", fontsize=14)
ax.set_ylabel("Average Total Bill ($)", fontsize=14)
ax.legend(title="Gender", loc="upper left", fontsize=10, title_fontsize=12)
plt.show()
```

##### Adjust Tick Labels

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=tips, x="day", y="total_bill", errorbar=None, hue="sex")

ax.set_title("Average Total Bill by Day and Gender", fontsize=16, fontweight='bold')
ax.set_xlabel("Day of the Week", fontsize=14)
ax.set_ylabel("Average Total Bill ($)", fontsize=14)
ax.legend(title="Gender", loc="upper right", fontsize=10, title_fontsize=12)

plt.xticks(fontsize=12) # Set font size
plt.xticks(rotation=45) # Rotate x-axis labels
plt.yticks(fontsize=12)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load built-in Seaborn dataset
tips = sns.load_dataset("tips")

# Set Seaborn theme and color palette
sns.set_style("darkgrid")
sns.set_palette("Set2")

# Create a new figure
plt.figure(figsize=(10, 6), dpi=100)

# Create a barplot with Seaborn
ax = sns.barplot(data=tips, x="day", y="total_bill", hue="sex", errorbar=None)

# Matplotlib customizations
ax.set_title("Average Total Bill by Day and Gender", fontsize=16, fontweight='bold', color='navy')
ax.set_xlabel("Day of the Week", fontsize=14)
ax.set_ylabel("Average Total Bill ($)", fontsize=14)
ax.legend(title="Gender", loc="upper right", fontsize=10, title_fontsize=12)

# Customize tick labels
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```
