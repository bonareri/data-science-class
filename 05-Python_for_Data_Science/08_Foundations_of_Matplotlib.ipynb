### Matplotlib Foundations

`Matplotlib` is a powerful Python library for data visualization. It allows you to create a variety of static, animated, and interactive plots.

Matplotlib is a multi-platform data visualization library built on NumPy arrays, and designed to work with the broader SciPy stack.

It was conceived by John Hunter in 2002, originally as a patch to IPython for enabling interactive MATLAB-style plotting via gnuplot from the IPython command line.

#### Why Use Matplotlib?
- To visualize patterns, trends, and relationships in data
- Works well with NumPy and Pandas
- Highly customizable (colors, labels, legends, etc.)
- Great for exploratory data analysis (EDA)

#### Basic Components of a Plot
No matter which library you use, most plots involve:

| Element         | Description                                        |
| --------------- | -------------------------------------------------- |
| `x-axis`        | Horizontal axis — usually the independent variable |
| `y-axis`        | Vertical axis — usually the dependent variable     |
| `data`          | What you're visualizing (numbers, categories)      |
| `labels`        | Axis titles, legend, and chart title               |
| `markers/lines` | How the data is drawn (dots, bars, lines, etc.)    |

#### What Are Variables in Data?
In a dataset or experiment, a `variable` is anything you measure or observe — like age, income, temperature, or test scores.

##### Independent vs Dependent Variables

| **Type**                    | **Definition**                                                                                   |
| --------------------------- | ------------------------------------------------------------------------------------------------ |
| **Independent Variable** | The variable you **change or control**. It causes a change in something else.                    |
| **Dependent Variable**   | The variable you **measure or observe**. It changes **in response to** the independent variable. |


#### Real-World Examples

| **Scenario**                                   | **Independent**         | **Dependent**            |
| ---------------------------------------------- | ----------------------- | ------------------------ |
| Studying hours vs. exam score                  | Hours studied           | Exam score               |
| Temperature vs. ice cream sales                | Temperature             | Ice cream sales          |
| Advertising budget vs. product sales           | Advertising spend       | Product sales            |
| GDP per capita vs. Life expectancy (Gapminder) | GDP per capita (x-axis) | Life expectancy (y-axis) |

##### In Plots:
- The `independent` variable usually goes on the `x-axis` (horizontal)
- The `dependent` variable usually goes on the `y-axis` (vertical)

### General Matplotlib Tips

Before we dive into the details of creating visualizations with Matplotlib, there are a few useful things you should know about using the package

In Matplotlib:

### Single Variable (Univariate)

Bar Chart - Sales by Category

##### MATLAB-style Interface

When you use `plt`, you’re working on the current active plot implicitly.

The MATLAB-style tools are contained in the pyplot (plt) interface.

```python
sales_by_category = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

sales_by_category.plot(kind = 'bar', color = 'green')

plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')

plt.xticks(rotation=45)

plt.show()
```

This works fine because:
- `pandas.plot()` internally creates and manages the figure/axes
- `plt` applies changes to the `current axes`

Limitation:
- You don’t have an explicit reference to the axes
- Harder to manage multiple subplots or complex figures

#### Object-Oriented style (ax)

```python
sales_by_category

fig, ax = plt.subplots(figsize=(8, 6))

ax = sales_by_category.plot(kind='bar', color='green', ax=ax)

ax.set_title('Total Sales by Category', fontsize=16, fontweight='bold')
ax.set_xlabel('Category')
ax.set_ylabel('Total Sales')

plt.xticks(rotation=45)
plt.show()
```

Advantages:
- You directly control that specific plot
- Better for multiple charts
- Cleaner and more professional

### Sales by Segment - Pie Chart

```python
sales_by_segement = df.groupby('Segment')['Sales'].sum().sort_values(ascending=False)

total = sales_by_segement.sum()
percentages = (sales_by_segement / total) * 100

# build custom labels
labels = [f"{seg} ({pct}%) " for seg, pct in zip(sales_by_segement.index, percentages.round(1))]

sales_by_segement.plot(
    kind = 'pie',
    startangle=90,
    labels=labels,
    colors=['#ff9999','#66b3ff','#99ff99']
)

plt.title('Total Sales by Segment')

plt.show()
```

### Top 10 Customers by Sales

```python
import matplotlib.cm as cm
import matplotlib.colors as mcolors

# Top 10 customers by sales
top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)

# Normalize sales for color mapping
norm = mcolors.Normalize(vmin=top_customers.min(), vmax=top_customers.max())
colors = cm.viridis(norm(top_customers.values))

# Plot the visualization
ax = top_customers.plot(
    kind = 'barh',
    color = colors,
    figsize=(10, 6),
)

ax.set_title('Top 10 Customers by Sales', fontsize=16, fontweight='bold')
ax.set_xlabel('Total Sales')
ax.set_ylabel('Customer Name')
plt.show()
```

#### What is Plotly?
`Plotly` is an open-source Python library used to create interactive and visually appealing data visualizations. It’s especially powerful for:

- Exploratory data analysis (EDA)
- Dashboards
- Web-based visualizations

Unlike static plots in Matplotlib or Seaborn, Plotly plots are fully interactive — you can zoom, hover, filter, and export directly from the chart.

#### Line Chart — Life Expectancy Over Time

- `x='year'`: time on the horizontal axis.
- `y='lifeExp'`: measured outcome (dependent variable).
- `color='country'`: different lines for each country.

```python
df_east_africa = df[df['country'].isin(['Kenya', 'Uganda', 'Tanzania'])]

fig = px.line(df_east_africa, x='year', y='lifeExp', color='country', title = 'Life Expectancy Over Time')

fig.show()
```

When you call:

`fig.show()`

Plotly uses its internal display engine (based on plotly.io) to render the figure as rich HTML/JavaScript inside the notebook output cell. This is what makes the chart interactive.

To do this, Plotly relies on:
- IPython (to display rich output)
- `nbformat` (to properly format and render notebook MIME types like HTML)

#### Scatter Plot — Multivariate View

visualize multiple variables at once.

```python
df_2007 = df[df['year'] == 2007]

fig = px.scatter(df_2007, 
                 x = 'gdpPercap', y = 'lifeExp',
                 size = 'pop', color = 'continent',
                 log_x = True,
                 hover_name = 'country', 
                 title = 'GDP vs Life Expectancy (2007)'
                 )
fig.show()
```

- `x`: GDP per capita (independent variable)
- `y`: Life expectancy (dependent variable)
- `size`: population (adds a 3rd dimension)
- `color`: continent (grouping)
- `hover_name`: tooltip info (country)
- `log_x=True`: fixes skewed GDP values

#### Bar Chart — Total Population by Continent

```python
df_bar = df_2007.groupby('continent')['pop'].sum().reset_index().sort_values(by='pop', ascending=False)
fig = px.bar(df_bar, x = 'continent', y = 'pop', 
             title = 'Total population by Continent')
fig.show()
```

#### Box Plot — Life Expectancy Spread
distribution across groups.

```python
fig = px.box(df_2007, x='continent', y='lifeExp', points= "all", hover_name='country',
             title='Life Expectancy Distribution by Continent')
fig.show()
```

#### Choropleth Map — Global Life Expectancy
##### What is a Choropleth Map?
A `Choropleth Map` is a thematic map where regions (countries, states, etc.) are colored or shaded based on the value of a numerical variable.

Useful for:

- Visualizing geographic patterns
- Comparing values across locations
- Displaying country-level or region-level statistics

```python
fig = px.choropleth(df_2007, 
                    locations = 'iso_alpha',
                    color = 'lifeExp',
                    hover_name = 'country',
                    projection = 'natural earth',
                    title = 'Life Expectancy by Country'
)
fig.show()
```

Parameters

| Parameter                | Description                                                           |
| ------------------------ | --------------------------------------------------------------------- |
| `locations`              | Location ID (e.g., ISO codes like `"KEN"` for Kenya)                  |
| `color`                  | The numeric column used to determine color intensity                  |
| `hover_name`             | Column shown on hover                                                 |
| `color_continuous_scale` | Color theme (e.g., `'Viridis'`, `'Blues'`, `'Plasma'`, `'Reds'`)      |
| `projection`             | Map type (e.g., `"natural earth"`, `"equirectangular"`, `"mercator"`) |
| `title`                  | Chart title                                                           |

#### What is a Map Projection?
A `projection` is a method used to transform the 3D surface of the Earth (a sphere or ellipsoid) into a 2D flat map.

Since you can't perfectly flatten a globe, every projection has some distortion — in shape, size, distance, or direction.

#### Why Projections Matter in Choropleth Maps
When you use:

`projection="natural earth"`

You're telling Plotly how to `flatten` the Earth for your map display.

Different projections emphasize different features:

- Some keep area accurate (good for comparing regions)
- Some preserve shapes (good for navigation)
- Some minimize distortion near the equator or poles

Common Projections in Plotly

| Projection Type          | Description                                                              |
| ------------------------ | ------------------------------------------------------------------------ |
| `"natural earth"`        | A **compromise projection** (balanced, aesthetically pleasing)           |
| `"equirectangular"`      | Straightforward, but **distorts shape** at poles                         |
| `"mercator"`             | Preserves angles; used in **Google Maps** but **greatly enlarges poles** |
| `"orthographic"`         | Makes the Earth look like a **3D globe**                                 |
| `"azimuthal equal area"` | Good for **preserving area**, often used in scientific maps              |
| `"winkel tripel"`        | A modern compromise used by **National Geographic**                      |
