AREA PLOTS

Area plots are used to represent quantitative data visually. They are similar to line plots, but the area between the line and the axis is filled with color or shading.
Area plots are useful for showing the cumulative sum of several quantities over time or categories.

```python
import numpy as np

import matplotlib.pyplot as plt

 #Example 1: Simple Area Plot
x = np.arange(1, 6)
y = np.array([3, 5, 2, 8, 7])

plt.fill_between(x, y, color='black', alpha=1) # Fill the area under the curve
plt.plot(x, y, color='green', alpha=0.6, linewidth=2) # Plot the line on top
plt.title('Simple Area Plot') # Title of the plot
plt.xlabel('months') # X-axis label
plt.ylabel('sales') # Y-axis label
plt.show()

# Example 2: Stacked Area Plot
# Stacked area plots show how different groups contribute to the whole over time.
y1 = np.array([3, 4, 1, 6, 5])
y2 = np.array([1, 2, 5, 2, 1])
y3 = np.array([2, 1, 3, 1, 2])


plt.stackplot(x, y1, y2, y3, labels=['stools', 'desks', 'tables'], alpha=0.8)
plt.legend(loc='upper left')
plt.title('supermarket funiture')
plt.xlabel('furniture')
plt.ylabel('sales')
plt.show()

```

Notes:
# - Use `fill_between()` for a single area plot, and `stackplot()` for multiple series.
# - The `alpha` parameter controls transparency.
# - Area plots can be misleading if the data series overlap significantly, so use them when the cumulative sum is meaningful.
# - Always label your axes and provide a legend for clarity.

# Area plots are great for visualizing part-to-whole relationships and trends over time.
