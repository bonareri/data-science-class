Density Contour Plots
Density contour plots are powerful visualizations used to display the distribution of two-dimensional data. They are particularly useful for visualizing the "density" or concentration of data points in a scatter plot, where simply plotting all points might lead to overplotting in dense areas.

Key Concepts:
Kernel Density Estimation (KDE):

The core idea behind these plots is often Kernel Density Estimation (KDE). KDE is a non-parametric way to estimate the probability density function (PDF) of a random variable.

In a 2D context, it smooths the observed data points to create a continuous probability surface. Areas with higher concentrations of points will have higher estimated densities.

scipy.stats.gaussian_kde: This function from scipy is commonly used to perform KDE, assuming a Gaussian (normal) kernel.

Contour Lines and Filled Contours:

Contour Lines: These lines connect points of equal estimated density, similar to contour lines on a topographic map connecting points of equal elevation. They show the "shape" of the distribution.

Filled Contours: These fill the regions between contour lines with different colors, providing a more intuitive visual representation of varying density levels. This helps to quickly identify areas of high and low density.

```python
#filled density and contour plot
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

np.random.seed(0)
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)

xmin, xmax = x.min() - 1, x.max() + 1
ymin, ymax = y.min() - 1, y.max() + 1
xi, yi = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]

kde = gaussian_kde([x, y])
zi = kde(np.vstack([xi.flatten(), yi.flatten()]))
zi = zi.reshape(xi.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xi, yi, zi, levels=20, cmap='viridis', alpha=0.7)  # Density contour
plt.colorbar(label='Density')  # Color bar for density

contour_lines = plt.contour(xi, yi, zi, colors='black', linewidths=0.5)

plt.title('Density and Contour Plot of Data Distribution', fontsize=16)
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)

plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

```python
#contour plot
feature_x = np.arange(0, 50, 2)
feature_y = np.arange(0, 50, 3)

# Creating 2-D grid of features
[X, Y] = np.meshgrid(feature_x, feature_y)

fig, ax = plt.subplots(1, 1)

Z = np.cos(X / 2) + np.sin(Y / 4)

ax.contour(X, Y, Z)

ax.set_title('Contour Plot')
ax.set_xlabel('feature_x')
ax.set_ylabel('feature_y')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

feature_x = np.linspace(-5.0, 3.0, 70)
feature_y = np.linspace(-5.0, 3.0, 70)

[X, Y] = np.meshgrid(feature_x, feature_y) #2D grid

fig, ax = plt.subplots(1, 1)

Z = X ** 2 + Y ** 2

# plots filled contour plot
ax.contourf(X, Y, Z)

ax.set_title('Filled Contour Plot')
ax.set_xlabel('feature_x')
ax.set_ylabel('feature_y')

plt.show()
```
